# 9. MatplotLib和Pandas绘图
 - plt.plot(x,y):以x,y对应坐标画图
 - plt.figure：画布
 - fig.savefig：保存图片
 - plt.subplot(rows, cols, index)：总共有rows行，cols列，现在画第index个
 - 样式：点状'o'，虚线'--'
 - 颜色：color='blue'
 - 标签: 
   - plt.legend()与在plt.plot内的label='xx'
   - legend控制label显示效果，loc控制label的显示位置
 - 散点图：plt.scatter(x,y)

```python
import numpy as np
import matplotlib.pyplot as plt
# 把图画在本notebook中
%matplotlib inline 
```

```python
x = np.linspace(0,10,20)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,y, '--')
```

## 9.1 虚线样式

```python
# 生成画布
fig = plt.figure()
plt.plot(x,y,'--')
```

```python
fig.savefig('d:/1.png')
```

```python
plt.subplot(2,1,1)  # 画2行1列的第一个
plt.plot(x,np.sin(x), '--')

plt.subplot(2,1,2) # 画2行1列的第二个
plt.plot(x,np.cos(x))
```

## 9.2 点状样式

```python
plt.plot(x,np.sin(x), 'o')
```

## 9.3 控制颜色

```python
plt.plot(x,np.sin(x), 'o', color='red')
```

## 9.4 标签

```python
x = np.linspace(0,10, 50)
plt.plot(x,np.sin(x),'--',label='sin(x)')
plt.plot(x,np.cos(x), 'o',label='cos(x)')
# legend控制label显示效果，loc控制label的显示位置
plt.legend(loc=0)
```

```python
# plot 函数可定制的参数很多
x = np.linspace(0,10,30)
plt.plot(x,np.sin(x), '-p', color='red', markersize = '18', linewidth = 4, markeredgecolor='blue', markeredgewidth=4)
```

## 9.5 限制范围

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))
plt.xlim(2,8)
plt.ylim(0.2,0.8)
```

## 9.6 散点图

 - 线性图

```python
plt.scatter(x,np.sin(x),s=100, c='red')
```

```python
plt.style.use('classic')
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x,y, c=colors, s = sizes, alpha=0.5)
plt.colorbar();
```
