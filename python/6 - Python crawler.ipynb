```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = "https://sz.lianjia.com/zufang"
resp = req.get(url)
print(resp)
print(resp.text)

soup = bs(resp.text, "lxml")
print(soup)

```

```python
links_div = soup.find_all('div', class_="content__list--item")
links = [div.a.get('href') for div in links_div]
# print(links)
links, len(links)
# links[1].get('href')

# links_div = soup.find_all('div', class_='content__list--item--aside')
# print(links_div)
```

```python
import requests as req
from bs4 import BeautifulSoup as bs

def get_page(url):
    resp = req.get(url)
    soup = BeautifulSoup(resp.text, 'lxml')
    return soup

def get_links(link_url):
    get_page(link_url)
    links_div = soup.find_all('div', class_="content__list--item")
    links = [div.a.get('href') for div in links_div]
    return links

links = get_links(url)
links

flat_url = 'https://sz.lianjia.com/zufang/SZ2414418334960320512.html'
soup = get_page(flat_url)
```
